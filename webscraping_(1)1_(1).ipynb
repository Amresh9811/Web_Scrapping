```python
# Necessary Libraries
import requests  # For making HTTP requests
from bs4 import BeautifulSoup  # For parsing HTML
import pandas as pd  # For working with data in DataFrame format
```

```python

# URL of the webpage
url = "https://arxiv.org/abs/2406.05518"

# Send a GET request to the webpage
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Extract the author's name
authors_div = soup.find("div", class_="authors")
author = authors_div.get_text(strip=True) if authors_div else "Author Not Found"

# Extract the publication date
dateline_div = soup.find("div", class_="dateline")
publication_date = dateline_div.get_text(strip=True) if dateline_div else "Date Not Found"

# Extract the abstract
abstract_div = soup.find("blockquote", class_="abstract")
abstract = abstract_div.get_text(strip=True) if abstract_div else "Abstract Not Found"

# Create a DataFrame
data = {
    "Author": [author],
    "Publication Date": [publication_date],
    "Abstract": [abstract]
}
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
output_csv_path = "web_scraped_info.csv"
df.to_csv(output_csv_path, index=False)

print("Information extracted and saved to", output_csv_path)
```

##From Google Scholor

```python
url = "import requests"
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://scholar.google.com/scholar?hl=en&as_sdt=0%2C5&q=chemistry&btnG="
r = requests.get(url)
soup = BeautifulSoup(r.content, "html.parser")
r
```

```python
Authors = []  # Initialize an empty list to store the authors

# Find all div elements with the class 'gs_a'
product_boxes = soup.find_all("div", class_="gs_a")

# Loop through each found element
for box in product_boxes:
    # Extract and clean the text content (author's name)
    Author_name = box.text.strip()
    # Add the author's name to the Authors list
    Authors.append(Author_name)
```

```python
Authors
```

```python
Abstract = []  # Initialize an empty list to store the abstracts

# Find all div elements with the class 'gs_rs'
product_boxes = soup.find_all("div", class_="gs_rs")

# Loop through each found element
for box in product_boxes:
    # Extract and clean the text content (abstract)
    Abstract_a = box.text.strip()
    # Append the abstract text to the Abstract list
    Abstract.append(Abstract_a)

# Print the list of abstracts
print(Abstract)
```

```python
print(len(Authors))  ; print(len(Abstract))
```

```python
min_length = min(len(Authors), len(Abstract))

# Trim lists to the same length
Author = Authors[:min_length]
Abstract = Abstract[:min_length]
```

```python
import pandas as pd   # create data frame using extracted data
df11=pd.DataFrame({"Author":Author,"Abstract":Abstract})
df11
```
