```python
import requests
from bs4 import BeautifulSoup

# Define a dictionary mapping URLs to their respective div classes
url_class_map = {
    'https://arxiv.org/abs/1601.00002': 'class1',
    'https://papers.ssrn.com/sol3/papers.cfm?abstract_id=4857257': 'class2',
    'https://www.semanticscholar.org/paper/MEGA6%3A-Molecular-Evolutionary-Genetics-Analysis-Tamura-Stecher/4c2b018392eca7e54588ae6ed8505985c878d8ba': 'class3',
    # Add more URLs and their corresponding classes as needed
}

def extract_content(url, class_map):
    if url not in class_map:
        raise ValueError(f"No class mapping found for URL: {url}")

    # Get the class for the given URL
    div_class = class_map[url]

    # Fetch the webpage
    response = requests.get(url)
    response.raise_for_status()  # Raise an exception for HTTP errors

    # Parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract content based on the div class
    content = soup.find_all('div', class_=div_class)

    # Process and return the extracted content
    extracted_texts = [div.get_text(strip=True) for div in content]
    return extracted_texts

# Example usage:
urls = [
     'https://arxiv.org/abs/1601.00002',
    'https://papers.ssrn.com/sol3/papers.cfm?abstract_id=4857257',
    'https://www.semanticscholar.org/paper/MEGA6%3A-Molecular-Evolutionary-Genetics-Analysis-Tamura-Stecher/4c2b018392eca7e54588ae6ed8505985c878d8ba'
]

for url in urls:
    try:
        extracted_content = extract_content(url, url_class_map)
        print(f"Content from {url}:")
        for text in extracted_content:
            print(text)
        print()
    except Exception as e:
        print(f"Failed to extract content from {url}: {e}")
```
